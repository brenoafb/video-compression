```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'foreman_qcif.yuv'
width = 176
height = 144
```

```python
file = open(filename, 'rb')
```

```python
y_size = width * height
y_bytes = file.read(y_size)
y_bytes
```

```python
matrix = np.array(list(y_bytes), dtype=np.uint8).reshape(height, width)
plt.imshow(matrix)
```

```python
uv_size = width * height // 4
u_bytes = list(file.read(uv_size))
```

```python
matrix = np.array(list(u_bytes), dtype=np.uint8).reshape(height//2, width//2)
plt.imshow(matrix)
```

```python
v_bytes = list(file.read(uv_size))
v_bytes
```

```python
matrix = np.array(list(v_bytes), dtype=np.uint8).reshape(height//2, width//2)
plt.imshow(matrix)
```
